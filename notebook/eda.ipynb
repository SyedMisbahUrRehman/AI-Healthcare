```python
import numpy as np
import pandas as pd

Q1 = np.percentile(df['BMI'], 25, method='midpoint')
Q3 = np.percentile(df['BMI'], 75, method='midpoint')
IQR = Q3 - Q1
bounds = Q1 - 1.45*IQR, Q3 + 1.45*IQR

df['BMI'] = df['BMI'].clip(*bounds)
```

```python
Q1 = np.percentile(df['Diastolic BP'], 25, method='midpoint')
Q3 = np.percentile(df['Diastolic BP'], 75, method='midpoint')
IQR = Q3 - Q1
bounds = Q1 - 1.5*IQR, Q3 + 1.5*IQR

df['Diastolic BP'] = df['Diastolic BP'].clip(*bounds)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


target_col = 'Health' 

numerical_cols = df.select_dtypes(include=[np.number]).columns.drop(target_col, errors='ignore')

sns.set(style="whitegrid")

for col in numerical_cols:
    plt.figure(figsize=(10, 5))
    
    sns.kdeplot(
        data=df,
        x=col,
        hue=target_col,
        fill=True,
        alpha=0.5,
        common_norm=False,
        palette="Set2",
    )
    
    plt.title(f"Feature Distribution of {col} by {target_col}", fontsize=14, weight='bold')
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.legend(title=target_col)
    plt.grid(True, linestyle="--", alpha=0.4)
    plt.tight_layout()
    plt.show()
```
